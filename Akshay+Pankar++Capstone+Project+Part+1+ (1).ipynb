identifying all the data which holds more than 30% of nan's. If yes then will drop all such columns

```python

nan_percentage = df_X.isna().mean() * 100


columns_to_drop = nan_percentage[nan_percentage > 30].index.tolist()


df_cleaned = df_X.drop(columns=columns_to_drop)


print("Original DataFrame:")
print(columns_to_drop)

print("\nDataFrame after dropping columns with more than 30% NaN:")
print(df_cleaned)
```
